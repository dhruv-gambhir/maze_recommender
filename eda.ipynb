```python
import re

map_sizes = []
for i in range(len(map_data)):
    match = re.search(r'mapSize (\d+) (\d+)', map_data[i])
    if match:
        size = int(match.group(1)), int(match.group(2))
        map_sizes.append(size)

print(map_sizes)
```

```python
import re

map_spawn = []
for i in range(len(map_data)):
    match = re.search(r'SpawnPoint (\d+) (\d+) (\d+)', map_data[i])
    if match:
        size = int(match.group(1)), int(match.group(2)), int(match.group(3))
        map_spawn.append(size)

print(map_spawn)
```

```python
map_spawn_table = pd.DataFrame(map_spawn, columns=['x', 'y', 'z'])

print(map_spawn_table)
```

```python
import re

map_end = []
for i in range(len(map_data)):
    match = re.search(r'EndPoint (\d+) (\d+) (\d+)', map_data[i])
    if match:
        size = int(match.group(1)), int(match.group(2)), int(match.group(3))
        map_end.append(size)

print(map_end)
```

```python
map_end_table = pd.DataFrame(map_end, columns=['x', 'y', 'z'])

print(map_end_table)
```

```python
distance = []
for i in range(len(map_end_table)):
    dist = int(np.sqrt((map_end_table.iloc[i]['x'] - map_spawn_table.iloc[i]['x'])**2 +
                   (map_end_table.iloc[i]['y'] - map_spawn_table.iloc[i]['y'])**2 +
                   (map_end_table.iloc[i]['z'] - map_spawn_table.iloc[i]['z'])**2))
    distance.append(dist)

# Create a new table with distances as the only column
distance_table = pd.DataFrame({'Distance': distance})

# Display the new table
distance_table
```

```python
def clean_data(data):
    return re.sub("[^a-zA-Z0-9; ]", "", data)
```

```python
new_maps['clean_data'] = new_maps['data'].apply(clean_data)
```

```python
def clean_description(description):
    if description is None:
        return ''
    return re.sub("[^a-zA-Z0-9 ]", "", description)
```

```python
new_maps['clean_description'] = new_maps['description'].apply(clean_description)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2))

tfdif_desc = vectorizer.fit_transform(new_maps['clean_description'])
```

```python
def search_desc(description):
    description = clean_description(description)
    query = vectorizer.transform([description])
    similarity = cosine_similarity(query, tfdif_desc).flatten()
    indices = np.argpartition(similarity, -30)[-30:]
    results = new_maps.iloc[indices][::-1]
    results = results.sort_values(by="difficulty", ascending=False)
    return results
```

```python
def find_similar_mazes(map_id):
    similar_users = merged[ (merged["id_x"] == map_id) & (merged["solved?"] == True)]
    similar_users_recs = merged[(merged["user_id_x"].isin(similar_users)) & (merged["difficulty"] >= 10)]

    x =similar_users_recs.value_counts("map_id") / len(similar_users)
    x_filtered = x.loc[x>= 0.01]

    all_users = merged[(merged["map_id"].isin(similar_users_recs["map_id"])) & (merged["difficulty"] >= 10)]
    y = all_users.value_counts("map_id") / len(all_users["user_id_x"].unique())

    rec_percentage = pd.concat([x_filtered, y], axis=1)
    rec_percentage.columns = ["similar", "all"]

    rec_percentage["score"] = rec_percentage["similar"] / rec_percentage["all"]

    rec_percentage = rec_percentage.sort_values(by="score", ascending=False)
    return rec_percentage.head(10).merge(maps, left_index = True, right_on="id")


    
```

```python
import ipywidgets as widgets
from IPython.display import display

map_input = widgets.Text(
    value='Enter description',
    description='Map description:',
    disabled=False
)

map_list = widgets.Output()

def on_type(datarey):
    map_list.clear_output()
    with map_list:
        title = datarey["new"]
        if len (title) > 0:
            display(search_desc(title)["description"][0:1])

map_input.observe(on_type, names="value")

display(map_input)
display(map_list)
```

```python
map_input_name = widgets.Text(
    value='Enter map',
    description='Map id:',
    disabled=False
)

recommendation_list = widgets.Output()

def on_type(datarey):
    recommendation_list.clear_output()
    with recommendation_list:
        title = datarey["new"]
        if len (title) > 0:
            display(find_similar_mazes(title)) 

map_input_name.observe(on_type, names="value")


display(map_input_name)
display(recommendation_list)

```
